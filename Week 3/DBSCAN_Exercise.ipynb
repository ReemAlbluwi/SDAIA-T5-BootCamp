#**DBSCAN**


#Step 1 : Import Libraries



```python
# prompt:  Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
```

#Step 2 : Prepare dataset
###**use make blobs with different numbers**



```python
# Load data in X
X, y_true = make_blobs(n_samples=500, centers=4,
                       cluster_std=0.50, random_state=0)
```

#Step 3 : Modeling The Data Using DBSCAN



```python
# Create DBSCAN model
dbscan = DBSCAN(eps=0.3, min_samples=5)

# Fit the model on the data
dbscan.fit(X)

# Obtain cluster labels
labels = dbscan.labels_

# Plot the results with distinct colors
plt.figure(figsize=(8, 6))
unique_labels = set(labels)
colors = plt.cm.get_cmap('rainbow', len(unique_labels))

for k, col in zip(unique_labels, colors(range(len(unique_labels)))):
    if k == -1:
        col = 'k'  # Black color for noise
    class_member_mask = (labels == k)
    xy = X[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=10, label=f'Cluster {k}' if k != -1 else 'Noise')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('DBSCAN Clustering on Blob Data with Clear Colors')
plt.legend()
plt.show()
```

#Step 4 : Evaluation Metrics For DBSCAN Algorithm In Machine Learning


```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, adjusted_rand_score

#-----------------------------------------------------------------------
X, y_true = make_blobs(n_samples=500, centers=4,
                       cluster_std=0.50, random_state=0)
dbscan = DBSCAN(eps=0.3, min_samples=5)

# Fit the model on the data
dbscan.fit(X)

# Obtain cluster labels
labels = dbscan.labels_

# Calculate Silhouette Coefficient
silhouette = silhouette_score(X, labels)
print(f"Silhouette Coefficient: {silhouette:.2f}")

# Calculate Adjusted Rand Index
ari = adjusted_rand_score(y_true, labels)
print(f"Adjusted Rand Index: {ari:.2f}")
```
