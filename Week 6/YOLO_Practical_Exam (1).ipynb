## Process Video Frames and Identify Vehicles Left the Road Form Each Street

For each frame in the video, use the YOLO model to detect and track vehicles. You will need to write a loop that processes each frame and applies the YOLO model.

```python
video_capture = cv2.VideoCapture('/content/YOLO_Video.mp4')
video_writer = cv2.VideoWriter('output_video.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 20, (int(video_capture.get(3)), int(video_capture.get(4))))
```

### counter code =)

```python
vehicle_counts = defaultdict(int)

line_1_start = (400, 200)
line_1_end = (400, 600)
line_2_start = (650, 1000)
line_2_end = (900, 1000)
line_3_start = (1600, 100)
line_3_end = (1600, 400)
line_4_start = (1000, 80)
line_4_end = (1300, 80)
```

```python
# object detection ===  car track
while True:
    ret, frame = video_capture.read()
    if not ret:
        break

    results = model(frame)

    for r in results[0].boxes.data.tolist():
        x1, y1, x2, y2, conf, cls = map(int, r[:6])

      # calcolate => car center
        cx = int((x1 + x2) / 2)
        cy = int((y1 + y2) / 2)


        if line_1_start[0] <= cx <= line_1_end[0] and cy > line_1_start[1]:
            vehicle_counts['top'] += 1
        elif line_2_start[0] <= cx <= line_2_end[0] and cy > line_2_start[1]:
            vehicle_counts['bottom'] += 1
        elif line_3_start[0] >= cx and line_3_start[1] <= cy <= line_3_end[1]:
            vehicle_counts['left'] += 1
        elif line_4_start[0] >= cx and line_4_start[1] <= cy <= line_4_end[1]:
            vehicle_counts['right'] += 1

        #draw box
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.circle(frame, (cx, cy), 5, (0, 0, 255), -1)
        #text = f"ID: {int(r[5])}, Class: {int(r[6])}, Conf: {round(r[4], 2)}"
       # cv2.putText(frame, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # draw line
    cv2.line(frame, line_1_start, line_1_end, (0, 255, 0), 2)
    cv2.line(frame, line_2_start, line_2_end, (255, 0, 0), 2)
    cv2.line(frame, line_3_start, line_3_end, (0, 0, 255), 2)
    cv2.line(frame, line_4_start, line_4_end, (255, 255, 0), 2)

    # display total and my name
    cv2.putText(frame, f"Top: {vehicle_counts['top']}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    cv2.putText(frame, f"Bottom: {vehicle_counts['bottom']}", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)
    cv2.putText(frame, f"Left: {vehicle_counts['left']}", (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
    cv2.putText(frame, f"Right: {vehicle_counts['right']}", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
    cv2.putText(frame, "Reem", (frame.shape[1] - 200, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

#----
    video_writer.write(frame)


video_capture.release()
video_writer.release()
cv2.destroyAllWindows()
```
